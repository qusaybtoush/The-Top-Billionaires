```python
#Importing the basic librarires fot analysis

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use("ggplot")  #using style ggplot

%matplotlib inline
import plotly.graph_objects as go
import plotly.express as px
```

```python
#Importing the dataset
df=pd.read_csv("../input/the-top-billionaires/The Top Billionaires.csv")

# look the data set
df.head()
```

```python
# looking the shape DataSet
df.shape
```

- it`s small data set

```python
# let see the data describe
df.describe().round(2)
```

# Visualizations 

```python
# library seaborn from make line graph

plt.figure(figsize=(20,7))
sns.lineplot(data=df,x="NAME",y="2018",label="2018")
sns.lineplot(data=df,x="NAME",y="2019",label="2019")
sns.lineplot(data=df,x="NAME",y="2020",label="2020")
sns.lineplot(data=df,x="NAME",y="2021",label="2021")
sns.lineplot(data=df,x="NAME",y="2022",label="2022")
```

```python
df.set_index("NAME").plot(kind="bar",figsize = (20,10))
```

- This graph show - Elon Musk - the first 3 year be less have money then he become rich person , he is genius ,

```python
plt.figure(figsize=(20,12))
plt.pie(data=df,x="2018",labels="NAME",autopct='%1.1f%%',shadow=True)
plt.title("The Rich in 2018")
plt.show()
```

#### In the year 2018 
- 1- Jeff Bezos
- 2- Bill Gates 

```python
plt.figure(figsize=(20,12))
plt.pie(data=df,x="2019",labels="NAME",autopct='%1.1f%%',shadow=True)
plt.title("The Rich in 2019")
plt.show()
```
